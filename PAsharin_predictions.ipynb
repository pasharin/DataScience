```python
def prepare_square(df):
    df.loc[df['Square'] > (df['Rooms']*50+25), 'Square'] = df['Rooms']*50+25
    df.loc[df['Square'] < (df['Rooms']*15), 'Square'] = df['Rooms']*15
    return df
```

```python
def prepare_KitchenSquare(df):
    df.loc[df['KitchenSquare'] < 5, 'KitchenSquare'] = df['Square']*0.20
    df.loc[df['KitchenSquare'] > 20, 'KitchenSquare'] = df['Square']*0.20
    return df
```

```python
data = prepare_KitchenSquare(data)
```

```python
def fillna_LifeSquare(df, df_source):
    df['LifeSquare'] = df['LifeSquare'].fillna(df['Square'] - df['KitchenSquare'])
    return df
```

```python
data = fillna_LifeSquare(data, data)
```

```python
def prepare_LifeSquare(df):
    df.loc[df['LifeSquare'] > df['Square'], 'LifeSquare'] = df['Square'] - df['KitchenSquare']
    df.loc[df['LifeSquare'] > (df['Square'] - df['KitchenSquare']), 'LifeSquare'] = df['Square'] - df['KitchenSquare']
    return df
```

```python
data = prepare_LifeSquare(data)
```

```python
data.loc[data['HouseFloor'] < data['Floor']]
```

```python
def prepare_HouseFloor(df):
    df.loc[df['HouseFloor'] < df['Floor'], 'HouseFloor'] = df['Floor'] + 1
    return df
```

```python
data = prepare_HouseFloor(data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = data[['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', 'Floor', 'mean_price_x']]
target = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression as LR, Ridge
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
model1 = LR()
```

```python
%%time
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_squared_error as mse
```

```python
print(mse(y_test, y_pred), r2(y_test, y_pred))
```

```python
model2 = Ridge(alpha = 0.05)
```

```python
%%time
model2.fit(X_train, y_train)
```

```python
y_pred_2 = model2.predict(X_test)
```

```python
print(mse(y_test, y_pred_2), r2(y_test, y_pred_2))
```

```python
model3 = RF(n_estimators=1000, max_depth=12, min_samples_leaf=5, random_state=42)
```

```python
%%time
model3.fit(X_train, y_train)
```

```python
y_pred_3 = model3.predict(X_test)
```

```python
print(mse(y_test, y_pred_3), r2(y_test, y_pred_3))
```
